```python
import pickle
import numpy as np
import pandas as pd
```

```python
# loading models and standard_scalar
model = pickle.load(open("model.pkl", "rb"))
standard_scalar = pickle.load(open("standardScaler.pkl", "rb"))
```

```python
# reading the evaluation dataset
eval_df = pd.read_csv("/content/AIQ1/aml_eval.csv")
```

```python
eval_df.head()
```

```python
drop_cols = ['TxnID', 'SenderID', 'ReceiverID', 'Unnamed: 13', 'token', 'time', 'Sender Location', 'Receiver Location']
```

```python
eval_df.drop(drop_cols,inplace=True, axis = 1)
```

```python
eval_df
```

```python
# one-hot encoding the trasaction types
transaction_type = pd.get_dummies(eval_df["TransactionType"])
transaction_type
```

```python
## adding the one-hot encoding into the dataset
eval_df_new = pd.concat([transaction_type,eval_df], axis = 1)
eval_df_new
```

```python
eval_df_new.drop("TransactionType", inplace = True, axis = 1)
```

```python
# standardizing the dataset
eval_df_std = pd.DataFrame(columns = eval_df_new.columns, data = standard_scalar.transform(eval_df_new))
eval_df_std
```

```python
## predicting using pre-trained model
pred = model.predict(eval_df_std)
pred[:10]
```

```python
## adding the result with the evaluation dataset
eval_df_with_result = pd.concat([eval_df, pd.Series(pred)], axis = 1)
eval_df_with_result
```

```python
# changing the last column name
eval_df_with_result.columns = eval_df_with_result.columns.astype(str)
eval_df_with_result.rename(columns = {"0" : "isSuspicious"},inplace=True)
eval_df_with_result
```
